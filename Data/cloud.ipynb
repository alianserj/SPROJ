```python
from itslive import velocity_cubes
import pandas as pd
import xarray as xr

# 1. Define your points (Lon, Lat)
coords = [
    (76.50, 35.75),  # Baltoro
    (77.00, 35.50),  # Siachen
    (74.50, 36.20)   # Hunza Region
]
names = ["Baltoro", "Siachen", "Hunza"]

print(f"Querying cloud for {len(coords)} glaciers...")

try:
    # 2. Fetch Time Series
    # The function returns a LIST of datasets [Dataset1, Dataset2, Dataset3]
    results_list = velocity_cubes.get_time_series(
        points=coords, 
        variables=["v", "date_dt"]
    )
    
    print("Cloud query successful! Processing data...")

    # 3. Process the List (The Fix)
    all_dfs = []
    
    # Loop through the results and the names together
    for i, ds in enumerate(results_list):
        # Convert this single glacier's dataset to a pandas DataFrame
        df_single = ds.to_dataframe().dropna().reset_index()
        
        # Add a label so we know which glacier this is
        df_single['Glacier_Name'] = names[i]
        
        # Add to our pile
        all_dfs.append(df_single)

    # Combine all into one big table
    final_df = pd.concat(all_dfs)

    print("\n--- Success! Data Retrieved ---")
    print(final_df.head())
    
    # Optional: Save to CSV
    # final_df.to_csv("my_shortlisted_glaciers.csv", index=False)

except Exception as e:
    print(f"Error details: {e}")
```
